```python
import matplotlib.pyplot as plt
import numpy as np


mod = np.random.rand(10)
bal = np.random.rand(10)

fig,ax = plt.subplots()

ax.plot(mod, color="red", marker="o")
ax.set_ylabel("Modularity", color="red",fontsize=14)

ax2=ax.twinx()
ax2.plot(bal,color="blue",marker="o")
ax2.set_ylabel("Balance",color="blue",fontsize=14)

ax.set_xlabel(r"$\lambda$", fontsize = 14)
plt.show()

fig.savefig('mod-bal.eps', format='eps',bbox_inches='tight')
```

```python
from utils import *
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.lines as mlines



k = 6
g = 3

n = k * g * 5

adj_mat, fair_mat, clusters, groups = gen_kleindessner(n, k, g, 0.4, .3, .2, .1)

b = np.zeros((clusters.size, clusters.max() + 1))
b[np.arange(clusters.size), clusters] = 1
aff = b @ b.T + 0.2

filt = np.ones([n,n]) - np.eye(n)
aff = aff * filt

Gc = nx.from_numpy_array(aff, create_using=nx.Graph, parallel_edges=False)
posc = nx.spring_layout(Gc)

G0 = nx.from_numpy_array(adj_mat, create_using=nx.Graph, parallel_edges=False)
pos0 = nx.spring_layout(G0)

fig = plt.gcf()
fig.set_size_inches(12, 12)
nx.draw(G0, posc,node_color=clusters,with_labels=False, node_size=0, width=0.05)


pos = [posc[i] for i in range(n)]
pos = np.array(pos)

c = len(np.unique(groups))

axx = []

markers = ["o", "s", "d" ,"v" , "^",  "<", ">", "." , ","]
for cc in range(c):
    ind = np.where(groups==cc)
    axx.append(plt.scatter(pos[ind,0], pos[ind,1], c= clusters[ind],\
                           marker=markers[cc], s=120))
    
# plt.legend(axx,
#            ('male', 'female'),
#            scatterpoints=1,
#            loc='lower left',
# #            ncol=3,
#            fontsize=12)  



blue_star = mlines.Line2D([], [], color='black', marker='o', linestyle='None',
                          markersize=16, label='Female')
red_square = mlines.Line2D([], [], color='black', marker='s', linestyle='None',
                          markersize=16, label='Male')


plt.legend(handles=[blue_star, red_square], ncol=c)

```
