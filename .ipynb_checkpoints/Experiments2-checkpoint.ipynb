### 1) Grid-search on $\lambda$ for different $Pr(i), n$ with equal-sized clusters
-  block sizes are equal; compatible with Assumption 10 in Theorem 1: $|V_s \cap C_l|= \frac{n}{kh}, s\in \left[h\right], l\in \left[k\right]$  
-  ùúÜ Parameter setup: $\lambda \in [0, \frac{1}{1000}, \frac{1}{100}, \frac{1}{10}, \frac{1}{2}, 1, 1.25, 1.5, 2, 10, 100]$ 
-  $k=5, h \in \{2, 5, 10\} $, 
-  $\bigg\{a=0.4, b=0.3, c=0.2, d=0.1\bigg\}, \bigg\{a=0.2, b=0.12, c=0.1, d=0.05\bigg\}, \bigg\{a,b,c,d\sim \Big(\frac{\log(n)}{n}\Big)^{\frac{2}{3}}\bigg\}$  
-  $n \in \bigl[2000, 4000, \ldots, 10000\bigr]$, 

###  Group Fair NMF

#### $\min_{{H}}\lVert{A}-{HH}^\top\rVert_F^2+\lambda\lVert{F}^\top{H}\rVert_F^2\quad \text{s.t.}\quad{H}\geq0$

```python
n = 100
h = 10


A = torch.rand(n,n)
F = torch.rand(n,h-1)

k = 5
lam = 1

H = torch.rand(n,k)

FF = F @ F.T
FFp = (torch.abs(FF) + FF) / 2
FFn = (torch.abs(FF) - FF) / 2

iter = 300
err = torch.zeros(iter)
for t in range(iter):
    
    Hu = A @ H       + (lam/2) * (FFn @ H)
    Hd = H @ H.T @ H + (lam/2) * (FFp @ H)
    H = H * (Hu/torch.maximum(Hd, eps)) ** 0.25
    
    err[t] = torch.norm(A - H @ H.T) ** 2 + lam * torch.norm(F.T @ H) ** 2
    
plt.plot(err)

pred = torch.argmax(H,dim=1)
pred
```

###  Group and Individual Fair NMF

#### $\min_{{H}}\lVert{A}-{HH}^\top\rVert_F^2+\lambda_1\lVert{F}^\top{H}\rVert_F^2+\lambda_2\lVert{RQ}{H}\rVert_F^2\quad \text{s.t.}\quad{H}\geq0$

```python
n = 100
h = 10


A = torch.rand(n,n)
F = torch.rand(n,h-1)
R = torch.rand(n,n)



k = 5
lam1 = 1
lam2 = 0

H = torch.rand(n,k)

FF = F @ F.T
FFp = (torch.abs(FF) + FF) / 2
FFn = (torch.abs(FF) - FF) / 2

Q = torch.eye(n)-(1/k) * torch.ones([n,n])

RQ0 = R @ Q
RQ  = RQ0.T @ RQ0
RQp = (torch.abs(RQ) + RQ) / 2
RQn = (torch.abs(RQ) - RQ) / 2

iter = 300
err = torch.zeros(iter)
for t in range(iter):
    
    Hu = A @ H       + (lam1/2) * (FFn @ H) + (lam2/2) * (RQn @ H)
    Hd = H @ H.T @ H + (lam1/2) * (FFp @ H) + (lam2/2) * (RQp @ H)
    H = H * (Hu/torch.maximum(Hd, eps)) ** 0.25
    
    err[t] = torch.norm(A - H @ H.T) ** 2 + lam1 * torch.norm(F.T @ H) ** 2\
     + lam2 * torch.norm(R @ Q @ H) ** 2
    
plt.plot(err)

pred = torch.argmax(H,dim=1)
pred
```

## Vanilla Symmetric NMF
### $\min_{{H}}\lVert{A}-{HH}^\top\rVert_F^2$

```python
df_NMF = pd.DataFrame(columns=['nodes (n)','clusters (k)','groups (h)','a','b','c','d','error','accuracy',
                           'average balance','cluster balances','Execution time'])

for j in range(len(n)):
    
    # calculate logarithmic probablity of edges a,b,c,d based on size of n
    logarithmic = (np.log(n[j])/n[j])**(2/3)
    multiplier = [10, 7, 4, 1]
    a[-1] = multiplier[0]* logarithmic
    b[-1] = multiplier[1]* logarithmic
    c[-1] = multiplier[2]* logarithmic
    d[-1] = multiplier[3]* logarithmic
    
    for param in range(len(a)):

        iter = 500
        err = torch.zeros(iter)
        for grps in h:
            
            # create sbm graph
            adj_mat, fair_mat, clusters, groups = gen_kleindessner(n[j], k, grps, a[param], b[param], c[param], d[param])
            # adjacency matrix 
            A = torch.tensor(adj_mat, dtype=torch.float)
            # the membership degree matrix H initialization
            H = torch.rand(n[j], k, dtype=torch.float)

            st = time.time()
            for t in range(iter):
                Hu = A @ H
                Hd = H @ H.T @ H
                H = H * (Hu / torch.maximum(Hd, eps)) ** 0.25

            et = time.time()
            elapsed_time = et - st
            
            pred = torch.argmax(H, dim=1)

            # transform fuzzy memberships (overlapping clusters) to strict (disjoint)
            pred_np = pred.numpy()

            # convert predicted labels to clusters 
            min_mistakes, reflow_true, aligned_clusters = align_clusters(clusters, pred_np)
            # calculate balance and error 
            balances, avg_balance = compute_group_balance(aligned_clusters.astype(int), groups)

            error = min_mistakes/n[j]
            acc= 1 - error
            
            # row of results to be appended to df
            results = dict()
            results = {'nodes (n)': n[j], 'clusters (k)': k, 'groups (h)': grps, 'a':a[param],'b':b[param],'c':c[param],
                       'd':d[param], 'error': error, 'accuracy': acc, 'average balance': avg_balance, 
                       'cluster balances': balances, 'Execution time':elapsed_time}
            temp = pd.DataFrame(pd.Series(results)).transpose() 
            #df = pd.concat([df, temp], ignore_index=True)
            df_NMF = df_NMF.append(temp, ignore_index=True)
            display(temp)

display(df_NMF)            
df_NMF.to_csv('vanilla_SymNMF_SBM' + '.csv', index=False)
```

## Vanilla SC-kmeans
-  $\min_{{H \in \mathbb{R}^{n\times k}}}\DeclareMathOperator{\Tr}{Tr}\big(H^\top LH\big) \quad s.t. \quad  H^\top H = I_k$
-  Then apply k-means to the rows of H that contains as its columns, orthonormal eigenvectors corresponding to the k smallest eigenvalues of L where 
-  $L = D-A$ and $A$ is the adjacency matrix

```python
df_SC = pd.DataFrame(columns=['nodes (n)','clusters (k)','groups (h)','a','b','c','d','error','accuracy',
                           'average balance','cluster balances','Execution time'])

for j in range(len(n)):
    
    # calculate logarithmic probablity of edges a,b,c,d based on size of n
    logarithmic = (np.log(n[j])/n[j])**(2/3)
    multiplier = [10, 7, 4, 1]
    a[-1] = multiplier[0]* logarithmic
    b[-1] = multiplier[1]* logarithmic
    c[-1] = multiplier[2]* logarithmic
    d[-1] = multiplier[3]* logarithmic
    
    for param in range(len(a)):

        iter = 500
        err = torch.zeros(iter)
        for grps in h:

            adj_mat, fair_mat, clusters, groups = gen_kleindessner(n[j], k, grps, a[param], b[param], c[param], d[param])

            st = time.time()

            pred = normal_sc(adj_mat, k)
            
            et = time.time()
            elapsed_time = et - st

            # convert predicted labels to clusters 
            min_mistakes, reflow_true, aligned_clusters = align_clusters(clusters, pred)
            # calculate balance and error 
            balances, avg_balance = compute_group_balance(aligned_clusters.astype(int), groups)

            error = min_mistakes/n[j]
            acc= 1 - error
            
            # row of results to be appended to df
            results = dict()
            results = {'nodes (n)': n[j], 'clusters (k)': k, 'groups (h)': grps, 'a':a[param],'b':b[param],'c':c[param],
                       'd':d[param], 'error': error, 'accuracy': acc, 'average balance': avg_balance, 
                       'cluster balances': balances, 'Execution time':elapsed_time}
            temp = pd.DataFrame(pd.Series(results)).transpose() 
            #df = pd.concat([df, temp], ignore_index=True)
            df_SC = df_SC.append(temp, ignore_index=True)
            display(temp)

display(df_SC)               
df_SC.to_csv('vanilla_SC_SBM' + '.csv', index=False)
```

-  $L = D-A$ and $A$ is the adjacency matrix

```python
df_FSC = pd.DataFrame(columns=['nodes (n)','clusters (k)','groups (h)','a','b','c','d','error','accuracy',
                           'average balance','cluster balances','Execution time'])

for j in range(len(n)):
    
    # calculate logarithmic probablity of edges a,b,c,d based on size of n
    logarithmic = (np.log(n[j])/n[j])**(2/3)
    multiplier = [10, 7, 4, 1]
    a[-1] = multiplier[0]* logarithmic
    b[-1] = multiplier[1]* logarithmic
    c[-1] = multiplier[2]* logarithmic
    d[-1] = multiplier[3]* logarithmic
    
    for param in range(len(a)):

        iter = 600
        err = torch.zeros(iter)
        for grps in h:

            adj_mat, fair_mat, clusters, groups = gen_kleindessner(n[j], k, grps, a[param], b[param], c[param], d[param])
            F = compute_F(groups)
            
            st = time.time()

            pred = group_fair_sc(adj_mat, F, k)
            
            et = time.time()
            elapsed_time = et - st
            
            # convert predicted labels to clusters 
            min_mistakes, reflow_true, aligned_clusters = align_clusters(clusters, pred)
            # calculate balance and error 
            balances, avg_balance = compute_group_balance(aligned_clusters.astype(int), groups)

            error = min_mistakes/n[j]
            acc= 1 - error
            
            # row of results to be appended to df
            results = dict()
            results = {'nodes (n)': n[j], 'clusters (k)': k, 'groups (h)': grps, 'a':a[param],'b':b[param],'c':c[param],
                       'd':d[param], 'error': error, 'accuracy': acc, 'average balance': avg_balance, 
                       'cluster balances': balances, 'Execution time':elapsed_time}
            temp = pd.DataFrame(pd.Series(results)).transpose() 
            #df = pd.concat([df, temp], ignore_index=True)
            df_FSC = df_FSC.append(temp, ignore_index=True)
            display(temp)

display(df_FSC)               
df_FSC.to_csv('Fair_SC_SBM' + '.csv', index=False)
```
